```python
# https://modelscope.cn/datasets/DAMO_NLP/yf_dianping
# 导入魔搭专属数据集加载工具 MsDataset，用来一键拉取 ModelScope 平台的公开数据集。
from modelscope.msdatasets import MsDataset

# 1. 加载完整数据集
full_ms_ds = MsDataset.load(
    "DAMO_NLP/yf_dianping",# 完整数据集 ID，格式命名空间/数据集名
    subset_name="default", # 数据集的子集名称，该数据集只有 default 这一个子集
    split="train" # 指定读取数据集的分片，此数据集原生只有 train 分片，代表读取全部原始训练数据。
)


# 调用内置方法，把魔搭数据集，转换成底层兼容的Hugging Face Dataset 标准对象。
# 方便后续切分训练 / 验证 / 测试集。
full_hf_ds = full_ms_ds.to_hf_dataset()

# 2. 第一次划分：总数据 90%训练临时集 + 10%测试集
# 随机种子，固定数字 = 固定随机打乱顺序，每次运行拆分结果一模一样，实验可复现
split1 = full_hf_ds.train_test_split(test_size=0.1, seed=42)
train_temp_hf = split1["train"]
test_hf = split1["test"]

# 3. 从临时训练集再切10%做验证集
split2 = train_temp_hf.train_test_split(test_size=0.1, seed=42)
train_hf = split2["train"]
val_hf = split2["test"]

# 打印数量 标准811划分
print(f"训练集train：{len(train_hf)}")
print(f"验证集val：{len(val_hf)}")
print(f"测试集test：{len(test_hf)}")
print("\n单条样本：", train_hf[0])
```
