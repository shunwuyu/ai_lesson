```python
# Python 数据处理的“瑞士军刀
# pandas 是每一个数据分析、金融、科研或 AI 新手都必须掌握的核心库
# 专门用来处理表格型数据（就像 Excel 表格），让你轻松完成：
# 读取、筛选、计算、绘图等操作。
# Series：一维数组（像一列数据）
# DataFrame：二维表格（像整个 Excel 表）
import pandas as pd
```

```python
# 创建一个简单的股票价格表
data = {
    '日期': ['2025-01-01', '2025-01-02', '2025-01-03'],
    '股票': ['宁德时代', '小米集团-W', '腾讯'],
    '收盘价': [200.5, 15.8, 805.0],
    '成交量(万)': [120, 300, 210]
}

df = pd.DataFrame(data)
# DataFrame 就是一个能让你像操作 Excel 表格一样处理数据的工具。
print(df)
```

```python
# 查看前两行
print(df.head(2))
```

```python
# 筛选“宁德时代”的记录
ningde = df[df['股票'] == '宁德时代']
print(ningde)
```

```python
high_price = df[df['收盘价'] > 200]
print(high_price)
```

```python
# 导入 matplotlib 的绘图模块
import matplotlib.pyplot as plt

# 假设我们有宁德时代连续几天的数据（模拟）
# 使用 pandas 生成从 2025-01-01 开始的连续 5 天的日期序列
dates = pd.date_range('2025-01-01', periods=5)
# 模拟的 5 天收盘价数据
prices = [200, 202, 198, 205, 210]
# 将日期和价格组合成一个 DataFrame，方便后续绘图
df_plot = pd.DataFrame({'日期': dates, '收盘价': prices})
# 设置“日期”为索引，并选取“收盘价”这一列进行绘图，生成折线图
# 默认折线图
df_plot.set_index('日期')['收盘价'].plot(title='宁德时代股价走势')
# 显示图表
plt.show()
```
