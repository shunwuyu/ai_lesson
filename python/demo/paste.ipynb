```python
# Python 本身没有像 Java/C 那样的内置数组
# list 是 Python 灵活有序、可修改的通用容器列表。
# 可存放任意类型数据，支持增删改查
# 长度可变，可随时增删元素，无需提前指定容量
L = ['Michael', 'Sarah', 'Tracy', 'Bob', 'Jack']
# 取前3个元素，应该怎么做？
# 笨办法：？ 取前N个元素就没辙了。
[L[0], L[1], L[2]]
# 取前N个元素，也就是索引为0-(N-1)的元素，可以用循环：
r = []
n = 3
# range(n) 生成 0 到 n-1 的整数序列，从 0 开始、左闭右开，一共 n 个数。
for i in range(n):
    r.append(L[i])
r

# Python提供了切片（Slice）操作符，能大大简化这种操作
# 取前3个元素，用一行代码就可以完成切片
# L[0:3]表示，从索引0开始取，直到索引3为止，但不包括索引3
L[0:3]
# 如果第一个索引是0，还可以省略：
L[:3]
# 也可以从索引1开始，取出2个元素出来：
L[1:3]
# 既然Python支持L[-1]取倒数第一个元素，那么它同样支持倒数切片，
# 记住倒数第一个元素的索引是-1。
L[-2:]
# 切片操作十分有用。我们先创建一个0-99的数列：
L = list(range(100))
L
# 前10个数
L[:10]
# 后10个数：
L[-10:]
# 前10个数，每两个取一个：
L[:10:2]
# 所有数，每5个取一个：
L[::5]
# 甚至什么都不写，只写[:]就可以原样复制一个list：
L[:]
```

```python
# 字符串'xxx'也可以看成是一种list，每个元素就是一个字符。
'ABCDEFG'[:3]
'ABCDEFG'[::2]
```

```python
# 实现一个trim()函数
def trim(s):
    return s.strip()

print(trim("  hello world  "))
```

```python
# 不用strip
# 利用切片操作，实现一个trim()函数，去除字符串首尾的空格，
# 注意不要调用str的strip()方法：
# 先找到首尾非空格位置，再通过切片截取中间字符串，实现去除首尾空格

def trim(s):
    # 处理左侧空格
    left = 0
    while left < len(s) and s[left] == ' ':
        left += 1
    # 处理右侧空格
    right = len(s)
    while right > left and s[right - 1] == ' ':
        right -= 1
    # 切片返回结果
    return s[left:right]

print(trim("  hello world  "))
```

```python
from openai import OpenAI

# DeepSeek API 配置
client = OpenAI(
    api_key="你的 DeepSeek API Key",
    base_url="https://api.deepseek.com/v1",
)

# 模型固定用 deepseek-chat
COMPLETION_MODEL = "deepseek-chat"

prompt = """
Consideration product : 工厂现货PVC充气青蛙夜市地摊热卖充气玩具发光蛙儿童水上玩具

1. Compose human readable product title used on Amazon in english within 20 words.
2. Write 5 selling points for the products in Amazon.
3. Evaluate a price range for this product in U.S.

Output the result in json format with three properties called title, selling_points and price_range
"""

def get_response(prompt):
    response = client.chat.completions.create(
        model=COMPLETION_MODEL,
        messages=[{"role": "user", "content": prompt}],
        max_tokens=1024,
        temperature=0.0,
    )
    return response.choices[0].message.content

# 输出结果
print(get_response(prompt))
```
